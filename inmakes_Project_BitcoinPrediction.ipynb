# importing libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# load datasheet

```python
df=pd.read_csv('BitcoinPrice.csv')
```

```python
"""
Displays the first 5 rows of the DataFrame `df`.

This function is useful for quickly inspecting the structure and contents of a DataFrame, especially during the exploratory data analysis phase of a project.
"""
df.head()
```

```python
df.info()
```

```python
df.drop(['Date'],1,inplace=True)
```

```python
df.head()
```

# predicting 30days in future

```python
p_days=30
df['Prediction']=df[['Price']].shift(-p_days)
```

```python
df.head()
```

```python
df.tail()
```

# seperate X and Y

```python
X=np.array(df.drop(['Prediction'],1))
X=X[:len(df)-p_days]
```

```python
X.shape
```

```python
X
```

# creating Y as np array

```python
Y=np.array(df['Prediction'])
Y=Y[:-p_days]
```

```python
Y.shape
```

```python
Y
```

# split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
P_days_array=np.array(df.drop(['Prediction'],1))[-p_days:]
```

```python
print(P_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf=RandomForestRegressor(n_estimators=1000,random_state=1)
Rf.fit(X_train,y_train)
print('Random Forest Accuracy:{:2f}%)'.format(Rf.score(X_test,y_test)*100))
```

# prediction

```python
Rf_predict=Rf.predict(X_test)
print(Rf.predict)
```

```python
print(y_test)
```

# Model prediction for 30 days

```python
Rf_predict_30=Rf.predict(P_days_array)
print(Rf_predict)
```
